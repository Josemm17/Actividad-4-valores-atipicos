```python
#Importamos las librerias pandas, numpy y matplotlib respectivamente
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Carga desde un archivo .csv sin indice
data= pd.read_csv('copia_gastos_costos_21.csv')
```

```python
#Verificamos información del DataFrame
#data.info()
data.head()
```

```python
data.info()
```

```python
#Corroboramos valores nulos
valores_nulos=data.isnull().sum()
valores_nulos
```

```python
data1 =data.fillna(method="ffill")
data1
```

**PROCEDIMIENTO PARA ELIMINAR VALORES NULOS EN DATAFRAME**

```python
#Eliminamos columnas innecesarias
data2= data1.drop(['Unnamed: 0', "FECHA", "FOLIO", "UUID", "RFC", "PROVEEDOR", "TIPO GASTO", "DESCRIPCION", "MP", "TIPO", "STATUS","POLIZA" ], axis=1)
data2
```

```python
#Realizamos diagrama de caja o bigote de cada columna del dataframe
fig = plt.figure(figsize =(15, 8))
data2.plot(kind='box', vert=False)
plt.title("Valores Atípicos del Dataframe")
plt.show() #dibujamos el diagrama
```

**PROCEDIMIENTO "DESVIACIÓN ESTÁNDAR" PARA ELIMINAR OUTLIERS EN DATAFRAME**

```python
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data2
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers del Dataframe
outliers= data2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos y los outliers se convierten en nulos en el DataFrame
data3= data2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data3
```

```python
#Corroboramos valores nulos del dataframe
valores_nulos=data3.isnull().sum()
valores_nulos
```

```python
#Reemplazamos valores atípicos (nulos) del dataframe con "mean"
#Realizamos una copia del dataframe
data_clean=data3.copy()
data_clean=data_clean.fillna(round(data3.mean(),1))
data_clean
```

```python
#Corroboramos valores nulos del dataframe LIMPIO
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
# Unimos la columnas con el dataframe

columnas_obj=data1[['Unnamed: 0', "FECHA", "FOLIO", "UUID", "RFC", "PROVEEDOR", "TIPO GASTO", "DESCRIPCION", "MP", "TIPO", "STATUS","POLIZA"]]
Datos_limpios2021 = pd.concat([columnas_obj, data_clean], axis=1)
Datos_limpios2021
```
